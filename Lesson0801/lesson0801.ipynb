```python
# 列印
print(12, 12.0, "12.0", "abc", sep=",")
print(12, 12.0, "12.0", "abc", sep=" ", end=".\n")
```

```python
# 變數的數值, 數據類型
name = "cat"
chinese = 78
english = 89.0
math = 65.0
print(name, chinese, english, math)
print(type(name)) # 字串型態
print(type(chinese)) # 整數型態
print(type(math)) # 浮點數
```

```python
# 算術運算子
print(2 + 6)  # 2+6
print(2**3)  # 2的3次方
print(4 + 5 * 10)  # 先乘除後加減 54
print((4 + 5) * 10) #9*10
print(8 // 3) # 求商數 2
print(7 % 4) # 求餘數 3
```

random.randint(a, b) 是 random 模組中的一個函數，用來生成一個隨機整數。這個整數的範圍包括兩個邊界 a 和 b（即 a 和 b 也可以是生成的數字）。

```python
import random 
random.randint(3,10)
```

```python
# 比較運算子
2 < 4
3 == 3
3 != 3
```

```python
# if 判斷式 if ..else..
score = 61
if score > 60:
    print("達標")
else:
    print("未達標")
```

```python
# 串列
student = ["nameless", 75, 80, 85]
print(student)
print(student[0]) # 提取串列元素0
print(student[1]) # 提取串列元素1
print(student[2]) # 提取串列元素2
print(student[3]) # 提取串列元素3
print(type(student)) # 串列 型態
```

```python
# 可以用append,附加在最後一個元素之後
student.append(90)
print(student)
```

```python
# 移除最後一個元素
student.pop()
student
```

```python
# dictionary 詞典
# key對應value

student_dict = {
    "name": "nameless",
    "chinese": 92,
    "english": 82,
    "math": 79,
}
print(student_dict["name"])  # 印出key去找出value
print(type(student_dict))  # 辭典型態
```

```python
# 印出詞典對應的值 
student_dict.values()
# # 中括號 表示暫時的資料
```
